```python
from pandas import DataFrame, read_csv
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
df: DataFrame = read_csv('https://pycourse.s3.amazonaws.com/banknote_authentication.txt',
                         header=None,
                         names=['variance', 'skewness', 'curtosis', 'entropy', 'class'])
```

```python
xtreino, xteste, ytreino, yteste = train_test_split(df.drop('class', axis=1),
                                                    df['class'], test_size=0.3,
                                                    random_state=1)
```

```python
clf_KNN = KNeighborsClassifier(n_neighbors=5)
clf_KNN.fit(xtreino, ytreino)
predicte = clf_KNN.predict(xteste)

print(accuracy_score(yteste, predicte) * 100)
```

```python
clf_arvore = DecisionTreeClassifier(random_state=1)
clf_arvore.fit(xtreino, ytreino)
predicte = clf_arvore.predict(xteste)

print(accuracy_score(yteste, predicte) * 100)
```

```python
clf_floresta = RandomForestClassifier(max_depth=8, random_state=1)
clf_floresta.fit(xtreino, ytreino)
predicte = clf_floresta.predict(xteste)

print(accuracy_score(yteste, predicte) * 100)
```

```python
clf_floresta.feature_importances_
```

```python
clf_svm = SVC(gamma='auto',kernel='rbf', random_state=1)
clf_svm.fit(xtreino, ytreino)
predicte = clf_svm.predict(xteste)

print(accuracy_score(yteste, predicte) * 100)
```

```python
clf_mlp = MLPClassifier(hidden_layer_sizes=(2,),
                        solver='lbfgs',
                        random_state=1)
clf_mlp.fit(xtreino, ytreino)
predicte = clf_mlp.predict(xteste)

print(accuracy_score(yteste, predicte) * 100)
```
